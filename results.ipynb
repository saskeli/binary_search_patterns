```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set()
```

```python
int_df = pd.read_csv("interlieved.tsv", sep="\t")
```

```python
int_df.drop("q", axis=1, inplace=True)
```

```python
plt.figure(figsize=(16, 10))
sns.boxplot(data=int_df)
plt.ylim(0, 2000)
plt.ylabel("time (ns)")
plt.title("Interlieved queries")
plt.savefig("interlieved_E7-8890.png")
plt.show()
```

```python
spe_df = pd.read_csv("separate.tsv", sep="\t")
```

```python
plt.figure(figsize=(16, 10))
sns.boxplot(x="type", y="time", data=spe_df)
plt.ylim(0, 2000)
plt.ylabel("time (ns)")
plt.title("Separate queries")
plt.savefig("separate_E7-8890.png")
plt.show()
```
